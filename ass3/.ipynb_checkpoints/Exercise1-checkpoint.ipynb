```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset into a Pandas dataframe
df = pd.read_csv('dist.csv', delimiter=';', header=None, names=['x', 'y', 'class'])

# Define the decision boundary function
def decision_boundary(x):
    if x > 3.94:
        return 0.5 * (18 - 2*x - np.sqrt(-724 + 256*x - 16*x**2))
    else:
        return 0.071 * (174 - 22*x - np.sqrt(23123 - 6144*x + 288*x**2))
```

```python
# Generate a scatter plot of the data
plt.scatter(df[df['class']==0]['x'], df[df['class']==0]['y'], c='blue', label='Class 0', marker='.')
plt.scatter(df[df['class']==1]['x'], df[df['class']==1]['y'], c='red', label='Class 1', marker='.')
plt.plot(np.linspace(-6, 10, 100), [decision_boundary(x) for x in np.linspace(-6, 10, 100)], color='black', label='True Decision Boundary')
plt.legend()
plt.show()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['x', 'y']], df['class'], test_size=0.2, random_state=42)

# Train SVM models using three different kernels
# Linear kernel
svm_linear = SVC(kernel='linear')
svm_linear.fit(X_train, y_train)

# RBF kernel
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train, y_train)

# Polynomial kernel
svm_poly = SVC(kernel='poly')
svm_poly.fit(X_train, y_train)

# Evaluate the predictive performance of the models on the test set
print("Linear kernel accuracy:", svm_linear.score(X_test, y_test))
print("RBF kernel accuracy:", svm_rbf.score(X_test, y_test))
print("Polynomial kernel accuracy:", svm_poly.score(X_test, y_test))
```

```python
# Plot the decision boundaries of each model
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, svm in enumerate([svm_linear, svm_rbf, svm_poly]):
    axs[i].scatter(X_test['x'], X_test['y'], c=y_test, cmap='magma', marker='.', label='Test data')
    axs[i].scatter(X_train['x'], X_train['y'], c=y_train, cmap='flag', marker='.', label='Training data')
    xlim = axs[i].get_xlim()
    ylim = axs[i].get_ylim()
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 100), np.linspace(ylim[0], ylim[1], 100))
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    axs[i].contourf(xx, yy, Z, alpha=0.1)
    axs[i].contour(xx, yy, Z, colors='black', linewidths=0.5)
    axs[i].plot(np.linspace(-6, 10, 100), [decision_boundary(x) for x in np.linspace(-6, 10, 100)], color='green', label='True decision boundary')
    axs[i].set_title('Kernel: ' + svm.kernel)
    axs[i].legend()

plt.show()
```
