```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("bm.csv", names=['x', 'y', 'job'])

# Extract the features and labels
X = data.drop("job", axis=1)
y = data["job"]

# Set the random seed and create a random permutation of the labels
np.random.seed(7)
r = np.random.permutation(len(y))

# Select the first n_s samples and create a dataset
n_s = 5000
X_s, y_s = X.iloc[r[:n_s], :], y.iloc[r[:n_s]]

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_s)

# Create and train the SVM with a Gaussian kernel
gamma = 0.5
C = 20
clf = svm.SVC(kernel='rbf', gamma=gamma, C=C)
clf.fit(X_scaled, y_s)

# Compute the training error
y_pred = clf.predict(X_scaled)
train_error = 1 - accuracy_score(y_s, y_pred)

print("Training Error:", train_error)

# Compute the indices of the support vectors
support_indices = clf.support_

# Plot the decision boundary, the data and the support vectors
plt.figure(figsize=(10, 5))

# Plot the support vectors and the decision boundary
plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_s, cmap=plt.cm.Paired)
plt.scatter(X_scaled[support_indices, 0], X_scaled[support_indices, 1], s=80, facecolors='none', edgecolors='k')
xlim = plt.xlim()
ylim = plt.ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 100), np.linspace(ylim[0], ylim[1], 100))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.title("Support Vectors and Decision Boundary")

# Plot the data and the decision boundary
plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_s, cmap=plt.cm.Paired)
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.title("Data and Decision Boundary")

plt.show()

```
