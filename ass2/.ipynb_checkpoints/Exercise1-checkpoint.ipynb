```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
my_data = pd.read_csv('GPUbenchmark.csv', dtype=np.float32).values

# extract X and y
X = my_data[:, :-1]
y = my_data[:, -1]

# normalize X
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
# plot Xi vs y for each feature
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i, ax in enumerate(axs.flat):
    ax.plot(X[:, i], y, 'bo', markersize=3)
    ax.set_xlabel(f"X{i+1}")
```

```python
# normal equation
Xe = np.c_[np.ones((len(X), 1)), X]
b = np.linalg.inv(Xe.T @ Xe) @ Xe.T @ y

# predict benchmark result for test data
testData = np.array([2432, 1607, 1683, 8, 8, 256], dtype=np.float32)
testData = (testData - X.mean(axis=0)) / X.std(axis=0)
testData = np.insert(testData, 0, 1)
predicted = b @ testData
print(f"Predicted benchmark result for test data: {predicted:.2f}")

# cost of normal equation
J = ((Xe @ b - y) ** 2).sum() / len(X)
print(f"Cost of normal equation: {J:.2f}")
```

```python
# gradient descent
def gradientDescent(Xe, y, a, num_iters):
    b = np.zeros(Xe.shape[1])
    gd = []
    for i in range(num_iters):
        b -= (a / len(Xe)) * Xe.T @ (Xe @ b - y)
        new_cost = ((Xe @ b - y) ** 2).sum() / len(Xe)
        gd.append([i, new_cost])
    return gd, b

# (a) hyperparameters for gradient descent
gd, b = gradientDescent(Xe, y, 0.01, 1000)
gd_arr = np.asarray(gd)
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(gd_arr[:, 0], gd_arr[:, 1])
ax.set_xlabel("Iteration")
ax.set_ylabel("Cost")

# (b) predict benchmark result for test data using gradient descent
predicted = b @ testData
print(f"Predicted benchmark result for test data using gradient descent: {predicted:.2f}")

plt.show()
```
