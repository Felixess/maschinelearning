```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('microchips.csv', header=None, names=['Test1', 'Test2', 'Class'])

# Separate the data into OK and fail groups
ok_data = data[data['Class'] == 1]
fail_data = data[data['Class'] == 0]

# Define the feature matrix X and target vector y
X = data[['Test1', 'Test2']].values
y = data['Class'].values

# 1. Plot the data
plt.scatter(ok_data['Test1'], ok_data['Test2'], c='green', label='OK')
plt.scatter(fail_data['Test1'], fail_data['Test2'], c='red', label='Fail')
plt.xlabel('Test 1')
plt.ylabel('Test 2')
plt.legend()
plt.show()
```

```python
# 2. Gradient descent for quadratic model
def mapFeatures(X, degree):
    """Map features to polynomial terms of degree less than or equal to d"""
    m, n = X.shape
    out = np.ones((m, 1))
    for i in range(1, degree + 1):
        for j in range(i + 1):
            term = (X[:, 0] ** (i - j)) * (X[:, 1] ** j)
            term = term.reshape(m, 1)
            out = np.hstack((out, term))
    return out

def sigmoid(z):
    """Sigmoid function"""
    return 1 / (1 + np.exp(-z))

def costFunction(X, y, theta):
    """Compute the cost function"""
    m = len(y)
    h = sigmoid(X @ theta)
    J = (-1 / m) * (y.T @ np.log(h) + (1 - y).T @ np.log(1 - h))
    return J

def gradientDescent(X, y, theta, alpha, num_iters):
    """Gradient descent optimization"""
    m = len(y)
    J_history = []
    for _ in range(num_iters):
        h = sigmoid(X @ theta)
        theta -= (alpha / m) * (X.T @ (h - y))
        cost = costFunction(X, y, theta)
        J_history.append(cost)
    return theta, J_history
```

```python
# Map features to quadratic terms
X_quad = mapFeatures(X, degree=2)

# Initialize theta and set hyperparameters
initial_theta = np.zeros(X_quad.shape[1])
alpha = 0.01
num_iters = 1000

# Perform gradient descent
theta, J_history = gradientDescent(X_quad, y, initial_theta, alpha, num_iters)

# Plot cost function over iterations
plt.plot(range(num_iters), J_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

```python
# Plot decision boundary
x1 = np.linspace(-1, 1.5, 100)
x2 = np.linspace(-1, 1.5, 100)
xx1, xx2 = np.meshgrid(x1, x2)
Z = np.zeros(xx1.shape)

# Map features to quadratic terms for decision boundary
X_quad_plot = mapFeatures(np.c_[xx1.ravel(), xx2.ravel()], degree=2)

# Compute predictions for decision boundary
Z = X_quad_plot @ theta
Z = sigmoid(Z)
Z = Z.reshape(xx1.shape)

# Plot decision boundary with scatter plot
plt.contour(xx1, xx2, Z, levels=[0.5], colors='blue')
plt.scatter(ok_data['Test1'], ok_data['Test2'], c='green', label='OK')
plt.scatter(fail_data['Test1'], fail_data['Test2'], c='red', label='Fail')
plt.xlabel('Test 1')
plt.ylabel('Test 2')
plt.title(f"Training Errors: {sum(np.round(X_quad @ theta) != y)}")
plt.legend()
plt.show()
```

```python
# 4. Gradient descent for polynomial of degree five (d = 5)
# Map features to polynomial terms
X_poly = mapFeatures(X, degree=5)

# Initialize theta and set hyperparameters
initial_theta_poly = np.zeros(X_poly.shape[1])
alpha_poly = 0.01
num_iters_poly = 1000

# Perform gradient descent
theta_poly, J_history_poly = gradientDescent(X_poly, y, initial_theta_poly, alpha_poly, num_iters_poly)

# Plot cost function over iterations
plt.plot(range(num_iters_poly), J_history_poly)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

# Plot decision boundary for polynomial of degree five
x1_poly = np.linspace(-1, 1.5, 100)
x2_poly = np.linspace(-1, 1.5, 100)
xx1_poly, xx2_poly = np.meshgrid(x1_poly, x2_poly)
Z_poly = np.zeros(xx1_poly.shape)

# Map features to polynomial terms for decision boundary
X_poly_plot = mapFeatures(np.c_[xx1_poly.ravel(), xx2_poly.ravel()], degree=5)

# Compute predictions for decision boundary
Z_poly = X_poly_plot @ theta_poly
Z_poly = sigmoid(Z_poly)
Z_poly = Z_poly.reshape(xx1_poly.shape)

# Plot decision boundary with scatter plot
plt.contour(xx1_poly, xx2_poly, Z_poly, levels=[0.5], colors='blue')
plt.scatter(ok_data['Test1'], ok_data['Test2'], c='green', label='OK')
plt.scatter(fail_data['Test1'], fail_data['Test2'], c='red', label='Fail')
plt.xlabel('Test 1')
plt.ylabel('Test 2')
plt.title(f"Training Errors: {sum(np.round(X_poly @ theta_poly) != y)}")
plt.legend()
plt.show()
```
