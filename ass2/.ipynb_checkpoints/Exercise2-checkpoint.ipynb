```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the data from CSV
data = pd.read_csv('secret_polynomial.csv', header=None, names=['x', 'y'])

# Split the data into training and test sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Set up the plot
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Fit polynomial models of degrees 1 to 6 and plot the results
for i, ax in enumerate(axes.flatten()):
    degree = i + 1

    # Create the design matrix for the specified degree
    X_train = np.column_stack([train_data['x'] ** d for d in range(1, degree + 1)])
    X_test = np.column_stack([test_data['x'] ** d for d in range(1, degree + 1)])

    # Fit a linear regression model to the design matrix
    model = LinearRegression()
    model.fit(X_train, train_data['y'])

    # Evaluate the model on the test data
    test_score = model.score(X_test, test_data['y'])

    # Plot the training and test data
    ax.scatter(train_data['x'], train_data['y'], color='blue', label='Training Data')
    ax.scatter(test_data['x'], test_data['y'], color='red', label='Test Data')

    # Plot the fitted curve
    x_range = np.linspace(data['x'].min(), data['x'].max(), 100)
    X_range = np.column_stack([x_range ** d for d in range(1, degree + 1)])
    y_range_pred = model.predict(X_range)
    ax.plot(x_range, y_range_pred, color='green', label=f'Degree {degree} Fit (Test Score: {test_score:.2f})')

    # Set plot title and labels
    ax.set_title(f"Polynomial Degree: {degree}")
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()

# Show the plot
plt.tight_layout()
plt.show()


```
