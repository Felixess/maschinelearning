```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score

# Load the data
data = pd.read_csv('microchips.csv', header=None, names=['Test1', 'Test2', 'Class'])

# Separate the data into OK and fail groups
ok_data = data[data['Class'] == 1]
fail_data = data[data['Class'] == 0]

# Define the feature matrix X and target vector y
X = data[['Test1', 'Test2']].values
y = data['Class'].values
```

```python
# 1. Construct classifiers with different degrees and plot decision boundaries
degrees = range(1, 10)
fig, axs = plt.subplots(3, 3, figsize=(12, 12))

for i, degree in enumerate(degrees):
    # Map features to polynomial terms
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)

    # Construct logistic regression classifier with regularization
    clf = LogisticRegression(C=10000, solver='liblinear')
    clf.fit(X_poly, y)

    # Plot decision boundary
    xx1, xx2 = np.meshgrid(np.linspace(-1, 1.5, 100), np.linspace(-1, 1.5, 100))
    X_plot = poly.transform(np.c_[xx1.ravel(), xx2.ravel()])
    Z = clf.predict(X_plot)
    Z = Z.reshape(xx1.shape)

    axs[i // 3, i % 3].contour(xx1, xx2, Z, levels=[0.5], colors='blue')
    axs[i // 3, i % 3].scatter(ok_data['Test1'], ok_data['Test2'], c='green', label='OK')
    axs[i // 3, i % 3].scatter(fail_data['Test1'], fail_data['Test2'], c='red', label='Fail')
    axs[i // 3, i % 3].set_title(f"Degree {degree}")
    axs[i // 3, i % 3].set_xlabel('Test 1')
    axs[i // 3, i % 3].set_ylabel('Test 2')

plt.tight_layout()
plt.show()
```

```python
# 2. Construct classifiers with different degrees and regularization parameter C=1
fig, axs = plt.subplots(3, 3, figsize=(12, 12))

for i, degree in enumerate(degrees):
    # Map features to polynomial terms
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)

    # Construct logistic regression classifier with regularization
    clf = LogisticRegression(C=1, solver='liblinear')
    clf.fit(X_poly, y)

    # Plot decision boundary
    xx1, xx2 = np.meshgrid(np.linspace(-1, 1.5, 100), np.linspace(-1, 1.5, 100))
    X_plot = poly.transform(np.c_[xx1.ravel(), xx2.ravel()])
    Z = clf.predict(X_plot)
    Z = Z.reshape(xx1.shape)

    axs[i // 3, i % 3].contour(xx1, xx2, Z, levels=[0.5], colors='blue')
    axs[i // 3, i % 3].scatter(ok_data['Test1'], ok_data['Test2'], c='green', label='OK')
    axs[i // 3, i % 3].scatter(fail_data['Test1'], fail_data['Test2'], c='red', label='Fail')
    axs[i // 3, i % 3].set_title(f"Degree {degree}")
    axs[i // 3, i % 3].set_xlabel('Test 1')
    axs[i // 3, i % 3].set_ylabel('Test 2')

plt.tight_layout()
plt.show()
```

```python
# 3. Cross-validation to compare regularized and unregularized models
degrees = range(1, 10)
errors_regularized = []
errors_unregularized = []

for degree in degrees:
    # Map features to polynomial terms
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)

    # Regularized logistic regression
    clf_regularized = LogisticRegression(C=1, solver='liblinear')
    scores_regularized = cross_val_score(clf_regularized, X_poly, y, cv=5)
    errors_regularized.append(len(y) - (scores_regularized.mean() * len(y)))

    # Unregularized logistic regression
    clf_unregularized = LogisticRegression(C=1e6, solver='liblinear')
    scores_unregularized = cross_val_score(clf_unregularized, X_poly, y, cv=5)
    errors_unregularized.append(len(y) - (scores_unregularized.mean() * len(y)))

# Plotting the results
plt.plot(degrees, errors_regularized, label='Regularized')
plt.plot(degrees, errors_unregularized, label='Unregularized')
plt.xlabel('Degree')
plt.ylabel('# Errors')
plt.legend()
plt.show()
```
